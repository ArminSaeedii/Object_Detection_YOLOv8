```python
from ultralytics import YOLO
import cv2
from google.colab.patches import cv2_imshow  # Colab-specific display

# Load pretrained model
model = YOLO("yolov8n.pt")

# Run inference
results = model.predict("/content/lab-hero-square-1fe2f13fa943105fe2c521df43eeb11c.jpg", conf=0.5)

# Display results
for result in results:
    img = result.plot()  # Get annotated image array
    cv2_imshow(img)      # Colab-compatible display
```
